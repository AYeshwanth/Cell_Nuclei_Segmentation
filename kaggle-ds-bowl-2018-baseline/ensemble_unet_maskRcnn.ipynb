```python
# hs this combines the individual cell masks into a single image
def combineMasks(masks):
    images = np.transpose(masks, [2, 0, 1])
    bigIm = images[0,:,:].astype(np.uint8)
    
    bigIm[:,:] = 0
    for image in images:
        tempIm = image.astype(np.uint8)
        bigIm = np.maximum(bigIm, tempIm)
    
    return bigIm

#this displays a single mask either combined or individual
def showSingleMask(mask):
    _, ax = plt.subplots(1, figsize=(8,8))
    ax.imshow(mask,cmap="Blues")
    plt.show()
```

# Loading u-net

```python
# Define IoU metric
def mean_iou(y_true, y_pred):
    prec = []
    for t in np.arange(0.5, 1.0, 0.05):
        y_pred_ = tf.to_int32(y_pred > t)
        score, up_opt = tf.metrics.mean_iou(y_true, y_pred_, 2)
        K.get_session().run(tf.local_variables_initializer())
        with tf.control_dependencies([up_opt]):
            score = tf.identity(score)
        prec.append(score)
    return K.mean(K.stack(prec), axis=0)
```

```python
model = load_model('model-hsUNet-512_50ep.h5', custom_objects={'mean_iou': mean_iou})
model.summary()
```

## classifying test set images with u-net

```python
# U-Net will resize every image to the sizes below. Mask R-CNN will not be affected (it load the data itself using test_ids)
IMG_WIDTH = 512 # hs original is 128
IMG_HEIGHT = 512 # hs original is 128
IMG_CHANNELS = 3


# Get and resize test images
X_test = np.zeros((len(test_ids), IMG_HEIGHT, IMG_WIDTH, IMG_CHANNELS), dtype=np.uint8)
sizes_test = []

print('Getting and resizing test images ... ')
sys.stdout.flush()

for n, id_ in tqdm(enumerate(test_ids), total=len(test_ids)):
    path = TEST_PATH + id_
    img = imread(path + '/images/' + id_ + '.png')[:,:,:IMG_CHANNELS]
    
    # keep track of original image sizes
    sizes_test.append([img.shape[0], img.shape[1]])
    
    #downsample the image
    img = resize(img, (IMG_HEIGHT, IMG_WIDTH), mode='constant', preserve_range=True)
    
    X_test[n] = img
```

```python
#classify all images in test set
preds_test = model.predict(X_test, verbose=1)

# All masks are predicted at 512*512. Below converts them all to the original image size.
# Also these predictions are in probabilitiy (from 0 to 1). We need to convert these to binary using a threshold.
probToBinaryTreshold = 0.5
preds_test_upsampled = []
unet_preds_test_upsampled_thresholded = []

for i in range(len(preds_test)):
    #resize images from 512*512 to original size
    resizedImage = resize(np.squeeze(preds_test[i]), (sizes_test[i][0], sizes_test[i][1]), mode='constant', preserve_range=True)
    preds_test_upsampled.append(resizedImage)
    
    #convert images to binary using a 0.5 threshold (any pixel with prob < 0.5 is 0 above 0.5 is 1)
    unet_preds_test_upsampled_thresholded.append(resizedImage > probToBinaryTreshold)
```

```python
# Run length encoding form U-Net. 
def rle_encoding(x):
    dots = np.where(x.T.flatten() == 1)[0]
    run_lengths = []
    prev = -2
    for b in dots:
        if (b>prev+1): run_lengths.extend((b + 1, 0))
        run_lengths[-1] += 1
        prev = b
    return run_lengths

def prob_to_rles(x, cutoff=0.5):
    lab_img = label(x > cutoff)
    for i in range(1, lab_img.max() + 1):
        yield rle_encoding(lab_img == i)
```

```python
new_test_ids = []
rles = []
for n, id_ in enumerate(test_ids):
    rle = list(prob_to_rles(combinedPredictions[n]))
    rles.extend(rle)
    new_test_ids.extend([id_] * len(rle))
```
